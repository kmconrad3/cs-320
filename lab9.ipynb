```python
import geopandas
world = geopandas.read_file(geopandas.datasets.get_path('naturalearth_lowres'))
cities = geopandas.read_file(geopandas.datasets.get_path('naturalearth_cities'))

eu = world[world["continent"] == "Europe"]
ax = eu.plot(column="pop_est", cmap="Oranges", edgecolor="0.8", figsize=(8,8))
#orange color map, light gray border around countries, increase color size
geopandas.sjoin(cities, eu).plot(ax=ax)

ax.set_xlim(-25, 45)
ax.set_ylim(30, 80)

geopandas.sjoin(cities, eu).plot(ax=ax, color="black", markersize=3) #small black markers for cities

ax.set_axis_off()
```
